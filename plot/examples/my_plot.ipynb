```python
from netCDF4 import Dataset
import numpy as np
import sys
sys.path.append('/global/homes/z/zhangtao/my_utils/plot/')
import contour
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
fid = Dataset('/global/homes/z/zhangtao/tmp/20180316.DECKv1b_A1.ne30_oEC.edison.cam.h0.1997-01.nc')
ps = fid.variables['TS'][0,:,:]
lat = fid.variables['lat'][:]
lon = fid.variables['lon'][:]

print(ps.shape)
```

```python
fig, axes = plt.subplots(2,2,figsize=(16,9), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=180)})
contour.plot_2d_contour_by_array(fig, axes[0][0], ps, lat, lon, 'TS', 'K', 'TEMP')
contour.plot_2d_contour_by_array(fig,axes[0][1], ps, lat, lon, 'TS', 'K', 'TEMP')
contour.plot_2d_contour_by_array(fig,axes[1][0], ps, lat, lon, 'TS', 'K', 'TEMP')
contour.plot_2d_contour_by_array(fig,axes[1][1], ps, lat, lon, 'TS', 'K', 'TEMP')
plt.tight_layout()
```
